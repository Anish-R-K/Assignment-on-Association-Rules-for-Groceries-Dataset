```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
groceries=[]
with open ('groceries.csv')as f:
    groceries=f.read()
groceries
groceries=groceries.split('\n')
groceries
```

```python
groceries_list=[]
for i in groceries:
    groceries_list.append(i.split(','))
groceries_list
```

```python
all_groceries_list=[i for item in groceries_list for i in item]
all_groceries_list
```

```python
from collections import Counter
item_freq=Counter(all_groceries_list)
item_freq
```

```python
item_freq=sorted(item_freq.items(),key=lambda x:x[1])
item_freq
```

```python
freq=list(reversed([i[1] for i  in item_freq]))
freq
```

```python
item=list(reversed([i[0]for i in item_freq]))
item
```

```python
groceries_series=pd.DataFrame(pd.Series(groceries_list))
groceries_series=groceries_series.iloc[:9835,:]
groceries_series
groceries_series.columns=["transactions"]
groceries_series
```

```python
x=groceries_series.transactions.str.join(sep=',').str.get_dummies(sep=',')
x
```

```python
freq_items=apriori(x,min_support=0.005,max_len=4,use_colnames=True)
freq_items
```

```python
freq_items.sort_values('support',ascending=False,inplace=True)
```

```python
freq_items
```

```python
rules=association_rules(freq_items,metric='lift',min_threshold=1)
rules
```

```python
rules.sort_values('lift',ascending=False)
```
